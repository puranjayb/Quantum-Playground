```python
from qiskit import QuantumCircuit, execute, Aer
from qiskit.visualization import plot_histogram


qc = QuantumCircuit(1, 1) # one Qubit, 1 classical bit


qc.h(0) #superposition

qc.measure(0, 0) # store the result in the classical bit after measuring

backend = Aer.get_backend("qasm_simulator") #check for free backend
job = execute(qc, backend, shots=1000) #running the quantum circuit
result = job.result()
counts = result.get_counts(qc)

print(type(counts))


print(counts)
plot_histogram(counts)
```

```python
from qiskit import QuantumCircuit, execute, Aer
from qiskit.visualization import plot_histogram

qc = QuantumCircuit(4, 4)

for i in range(4):
    qc.h(i)

for i in range(4):
    qc.measure(i, i)

backend = Aer.get_backend('qasm_simulator')
job = execute(qc, backend, shots=100)
result = job.result()
counts = result.get_counts(qc)

print(counts)
plot_histogram(counts)

# binary_strings = list(counts.keys())
# for bstring in binary_strings:

#     random_numbers = list( int(bstring, 2) for bstring in binary_strings)

# print(random_numbers)

# while (random_number in random_numbers) > 9:
#     job = execute(qc, backend, shots=1)
#     result = job.result()
#     counts = result.get_counts(qc)
#     binary_string = list(counts.keys())[0]
#     random_number = int(binary_string, 2)

#     print(random_number)

# plot_histogram(counts)
```

```python
import random
import matplotlib.pyplot as plt

counts = {}

for i in range(100):
    n = random.randint(0, 15)
    binary_string = format(n, '04b')

    if binary_string in counts:
        counts[binary_string] += 1
    else:
        counts[binary_string] = 1

# for key, value in counts.items():
#     print(f'{key}: {value}')

plt.bar(counts.keys(), counts.values())

for(x, y) in zip(counts.keys(), counts.values()):
    plt.text(x, y, str(y), ha='center', va='bottom')


plt.xticks(rotation=90)
plt.show
```
